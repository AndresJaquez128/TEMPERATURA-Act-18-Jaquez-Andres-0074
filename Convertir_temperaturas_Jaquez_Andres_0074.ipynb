## 1. Fórmulas de Conversión

Para movernos entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:** Multiplicamos por $1.8$ (o $9/5$) y sumamos $32$.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit a Celsius:** Restamos $32$ y dividimos el resultado entre $1.8$.
    $$°C = \frac{°F - 32}{1.8}$$

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir $25°C$ a $°F$
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$.
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** $25°C$ equivalen a **$77°F$**.

### Ejemplo B: Convertir $95°F$ a $°C$
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$.
2.  **Sustituir el valor:** $°C = (95 - 32) / 1.8$.
3.  **Restar:** $95 - 32 = 63$.
4.  **Dividir:** $63 / 1.8 = 35$.
* **Resultado:** $95°F$ equivalen a **$35°C$**.

## 3. Código en Python

Este programa es interactivo y utiliza las funciones solicitadas para gestionar la lógica y el flujo del usuario.

```python
def convertir_CF():
    """Convierte de Celsius a Fahrenheit."""
    celsius = float(input("Ingrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f">>> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertir_FC():
    """Convierte de Fahrenheit a Celsius."""
    fahrenheit = float(input("Ingrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8

# Ejecución del programa
```

### Explicación del Código:
* **`while True`:** Mantiene el programa corriendo hasta que el usuario decida salir.
* **`input()`:** Captura la entrada del usuario como texto. Usamos `float()` para permitir decimales en la temperatura.
* **`if/elif/else`:** Evalúa qué camino tomar según la opción marcada.
* **`break`:** Rompe el ciclo infinito cuando se selecciona la opción 3.
* **`f"{valor:.2f}"`:** Formatea el resultado para mostrar solo 2 decimales, haciendo que la lectura sea más limpia.

````python
def convertir_CF():
    """Convierte de Celsius a Fahrenheit."""
    celsius = float(input("Ingrese los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f">>> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertir_FC():
    """Convierte de Fahrenheit a Celsius."""
    fahrenheit = float(input("Ingrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f">>> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    """Punto de entrada principal con el menú interactivo."""
    while True:
        print("\n--- SISTEMA DE CONVERSIÓN DE TEMPERATURA ---")
        print("1. Convertir grados centigrados a fahrenheit")
        print("2. Convertir grados fahrenheit a grados centigrados")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == '1':
            convertir_CF()

        elif opcion == '2':
            convertir_FC()

        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Programa realizado por Jaquez Andres NL = 27")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
````
